# Take only values that are different from nan

# gold medal for each country

```python
gold_medals['Region'].value_counts().reset_index(name = 'Medal').head(5) 
```

```python
totalGoldMedals = gold_medals['Region'].value_counts().reset_index(name = 'Medal').head(5) 
g = sns.catplot(x='Region', y='Medal', data=totalGoldMedals, height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of models")
```

# Rio of olympics

```python
max_year=athletes_df['Year'].max()
print(max_year)
teams_names = athletes_df[(athletes_df['Year'] == max_year) & (athletes_df['Medal'] == 'Gold')].Team
teams_names.value_counts().head(10)
```

```python
sns.barplot(x=teams_names.value_counts().head(20), y=teams_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('countrywise medals for the year 2016')
```
